```python
arr = [9, 5, 3, 8, 4]

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key

print("Insertion Sorted Array:", arr)
```

```python
import time
import random
import matplotlib.pyplot as plt

# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Selection Sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Benchmarking function
def benchmark_sort(sort_func, size):
    arr = [random.randint(1, 10000) for _ in range(size)]  # Generate random array
    start_time = time.time()  # Start timer
    sort_func(arr.copy())  # Sort the array using the sort function
    end_time = time.time()  # End timer
    return end_time - start_time  # Return time taken

# Define input sizes for benchmarking
sizes = [5, 10, 20, 100, 500, 1000, 2000, 5000, 10000, 20000]

# Lists to store times
times_insertion = []
times_selection = []
times_bubble = []

# Perform benchmarking for each size
for size in sizes:
    times_insertion.append(benchmark_sort(insertion_sort, size))
    times_selection.append(benchmark_sort(selection_sort, size))
    times_bubble.append(benchmark_sort(bubble_sort, size))

# Plotting the results
plt.plot(sizes, times_insertion, label='Insertion Sort', marker='o')
plt.plot(sizes, times_selection, label='Selection Sort', marker='o')
plt.plot(sizes, times_bubble, label='Bubble Sort', marker='o')
plt.xlabel('Input size (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.title('Benchmark of Sorting Algorithms')
plt.grid(True)
plt.show()
```
